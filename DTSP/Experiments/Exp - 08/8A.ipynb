#### Problem Statement: Design a digital FIR Low Pass Filter using Rectangular Window. The following specfications are:
#### Cut-off Frequency = 0.3pi
#### Length  = 51

```python
# Importing in-built libraries of python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
# Design of FIR LPF using Rectangular Window
N = 51 # Type - I
b = signal.firwin(N, cutoff = 0.3, window = 'rect', pass_zero = True)
a = 1
print(np.round(b, decimals = 2))
```

```python
W, h = signal.freqz(b,a)
h_db = 20 * np.log10(abs(h))
plt.figure (figsize = (10, 4))
plt.plot (W/max(W), h_db)
plt.grid()
plt.title ('Your Name Digital LPF', fontsize = 15)
plt.xlabel ('W', fontsize = 12)
plt.ylabel ('H(w)', fontsize = 12)
```
