```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unit_step_fun(a,n):
    unit_step = []
    for s in n:
        if s >= a:
            unit_step.append(1)
        else:
            unit_step.append(0)
    return unit_step
```

```python
a = int(input('Enter shift of unit step a = '))
L = int(input('Enter length of sample numbers L = '))
n = np.arange(-L, L+1)
unit_step = unit_step_fun(a,n)
```

```python
plt.stem(n, unit_step, 'r')
plt.xticks(n)
plt.yticks([0, 1])
plt.xlabel('Discrete Time Variable', fontsize = 12)
plt.ylabel('Output Response', fontsize = 12)
plt.title('Unit Step Function', fontsize = 15)
```

```python
def unit_ramp_fun(a,n):
    unit_ramp = []
    for s in n:
        if s >= a:
            unit_ramp.append(s - a)
        else:
            unit_ramp.append(0)
    return unit_ramp
```

```python
a = int(input('Enter shift of unit step a = '))
L = int(input('Enter length of sample numbers L = '))
n = np.arange(-L, L+1)
unit_ramp = unit_ramp_fun(a,n)
```

```python
plt.stem(n, unit_ramp, 'r')
plt.grid()
plt.xticks(n)
plt.yticks([0, 1, 2, 3, 4, 5])
plt.xlabel('Discrete Time Variable', fontsize = 12)
plt.ylabel('Output Response', fontsize = 12)
plt.title('Unit Ramp Function', fontsize = 15)
```
