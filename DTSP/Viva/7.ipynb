```python
# Importing in-built libraries of python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
cutoff_low = 0.3
cutoff_high = 0.6
```

```python
N = 11 # Type - I
b = signal.firwin(N, [cutoff_low, cutoff_high], window = 'black', pass_zero = False)
a = 1
print(np.round(b, decimals = 2))
```

```python
W, h = signal.freqz(b,a)
h_db = 20 * np.log10(abs(h))
plt.figure (figsize = (10, 4))
plt.plot (W/max(W), h_db)
plt.grid()
plt.title ('Om Digital LPF Blackman', fontsize = 15)
plt.xlabel ('W', fontsize = 12)
plt.ylabel ('H(w)', fontsize = 12)
```
